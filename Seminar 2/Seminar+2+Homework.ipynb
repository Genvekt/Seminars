__Задание 1.__   
Доказать $(\pmb{AB})^{\top} = \pmb{B}^{\top}\pmb{A}^{\top}$.

__Решение__  
Assume that $$ \pmb{A} = (n \times m) $$ and $$\pmb{B} = (m \times k) $$
Then $$\pmb{AB}^{\top} = ((n \times m)(m \times k))^{\top} = (n \times k)^{\top} = (k \times n) $$
Also $$\pmb{B}^{\top}\pmb{A}^{\top} = (m \times k)^{\top}(n \times m)^{\top} = (k \times m)(m \times n) = (k\times n)$$
That means $$\pmb{AB}^{\top} = \pmb{B}^{\top}\pmb{A}^{\top}$$

__Определение.__ Матрица $\pmb{M}$ называется ортогональной, если выполняется следующее тождество:
$$
\pmb{M M}^\top = \pmb{M}^\top \pmb{M} = \pmb{I}
$$  
где $\pmb{I}$ - единичная матрица. Другими словами, для ортогональной матрицы верно:  
$$
\pmb{M}^{-1} = \pmb{M}^\top
$$
